```python
# Function to read
def read():
    csv = input("path to your csv please: ") 
    data_from_csv = pd.read_csv(csv)
    return data_from_csv.head()
read()
```

```python
# Clean the data
def clean():
    data = read()
    #Remove the null values
    data = data.dropna(axis=1,how="any") 
    #Remove the duplicates
    #data = data.drop_duplicates(inplace='True')
    #Detect and Remove outliers
    z_scores = stats.zscore(data.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']))
    # Calculate the 1st and 99th percentile thresholds
    lower_threshold = stats.norm.ppf(0.01)
    upper_threshold = stats.norm.ppf(0.99)

    # Filter data based on z-score thresholds
    filtered_entries = (z_scores > lower_threshold) & (z_scores < upper_threshold)
    filtered_data = data[filtered_entries.all(axis=1)]    
```
